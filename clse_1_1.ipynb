# EJERCICIO VALOR FUTURO CON PYTON 

```python
#Ejercicio de datos al usuario trasformados a int
principal_vp = int(input("ingrese el valor del prestamo:"))
tasa_interes_i = int(input("ingrese la tasa de interes:"))
años_n = int(input("ingrese el numero de años:"))

#calculo del valor futuro (interes simple)
#Dividimos la tasa de interes entre 100 para convertirla a decimal
valor_futuro = principal_vp * (1 + (tasa_interes_i/ 100)) ** años_n

print("-" * 30)
print(f"Al final del {años_n} años, tu inversion valdra: ${valor_futuro:.2f}")
```
