```python
# 1. Bienvenida e Interacción Inicial - 
# Mostramos una línea superior decorativa
print('=====================================================') 
# Mostramos el título con emojis para que sea amigable
print('💼 💱 Calculadora Financiera de Divisas - COP → USD / EUR') 
# Mostramos una línea separadora para claridad
print('-----------------------------------------------------') 

# 2. Entradas de Datos (Inputs & Casting)
# Iniciamos un bloque try para capturar errores de entrada (ValueError)
try:
    # Pedimos el monto en COP y lo convertimos a float (casting)
    monto_cop = float(input('Ingrese el monto en Pesos Colombianos (COP): ')) 
    # Pedimos la TRM del Dólar (COP por 1 USD) y la convertimos a float
    trm_dolar = float(input('Ingrese la TRM del Dólar (COP por USD): ')) 
    # Pedimos la tasa del Euro (COP por 1 EUR) y la convertimos a float
    tasa_euro = float(input('Ingrese la tasa del Euro (COP por EUR): ')) 
    # Mostramos por qué usamos float() en finanzas (impreso para el estudiante)
    print('Nota: usamos float() porque permite manejar decimales (p. ej. centavos) en cálculos financieros.') 

    # 3. Procesamiento (Cálculos)
    # Convertimos COP a USD usando la fórmula básica: monto / TRM
    monto_usd = monto_cop / trm_dolar 
    # Convertimos COP a EUR usando la fórmula básica: monto / tasa_euro
    monto_eur = monto_cop / tasa_euro 

    # 4. Salida de Información (Outputs)
    # Encabezado visual de resultados con emojis
    print('\n=================== Resultados ===================') 
    # Mostramos el monto ingresado en COP con separador de miles y dos decimales
    print(f'🔸 Monto ingresado: COP {monto_cop:,.2f}') 
    # Mostramos el equivalente en USD usando f-string y formato :,.2f
    print(f'💵 Equivalente en USD: {monto_usd:,.2f} USD') 
    # Mostramos el equivalente en EUR usando f-string y formato :,.2f
    print(f'💶 Equivalente en EUR: {monto_eur:,.2f} EUR') 
    # Línea separadora final
    print('-----------------------------------------------------') 
    # Mensaje de cierre amigable con emoji
    print('✅ Conversión completada. Gracias por usar la calculadora 💡') 

# Manejo de errores: el usuario ingresó texto no numérico y float() falló
except ValueError:
    # Aviso claro con emojis para indicar entrada inválida
    print('\n⚠️🚫 Entrada inválida: por favor ingresa sólo números. Ejemplo válido: 12345.67') 
```
