```python
# FinVes - Aplicación consola para inversores colombianos

# Pedimos el nombre del usuario (cadena de texto)
nombre = input("Ingrese su nombre: ").strip()

# Pedimos la meta mensual necesaria para vivir y la convertimos a float para cálculos financieros
while True:
    try:
        meta_mensual = float(input("Ingrese su Meta de Libertad Financiera (monto mensual en COP): ").strip())
        break
    except ValueError:
        print("⚠️ Entrada inválida. Escribe un número válido, por ejemplo: 1500000.00")

# Guardamos el perfil del usuario en un diccionario llamado perfil_usuario
perfil_usuario = {"nombre": nombre, "meta_mensual": meta_mensual}

# Definimos una tupla fija con los activos disponibles
activos_disponibles = ("Bancolombia", "Ecopetrol", "S&P500", "Bonos del Estado")

# Creamos una lista vacía para almacenar las inversiones del usuario
portafolio = []

# Mensaje de bienvenida breve
print(f"\n¡Bienvenido(a) {perfil_usuario['nombre']}! Comencemos a gestionar tu portafolio.\n")

# Menú interactivo usando un ciclo while
while True:
    print("Menú FinVes:")
    print("1. Registrar nueva inversión")
    print("2. Ver Estado del Portafolio y Libertad Financiera")
    print("3. Salir")
    opcion = input("Seleccione una opción (1/2/3): ").strip()

    if opcion == "1":
        # Mostrar activos disponibles
        print("\nActivos disponibles:")
        for i, activo in enumerate(activos_disponibles, start=1):
            print(f"{i}. {activo}")

        # Selección del activo por número
        try:
            eleccion = int(input("Ingrese el número del activo que desea elegir: ").strip())
            if not 1 <= eleccion <= len(activos_disponibles):
                print("⚠️ Opción fuera de rango. Volviendo al menú.\n")
                continue
        except ValueError:
            print("⚠️ Entrada inválida. Debes ingresar un número entero. Volviendo al menú.\n")
            continue

        activo_seleccionado = activos_disponibles[eleccion - 1]

        # Capturar monto en COP y tasa anual (porcentaje), convertidos a float
        try:
            monto_cop = float(input(f"Ingrese el monto a invertir en {activo_seleccionado} (COP): ").strip())
            tasa_anual = float(input("Ingrese la tasa de rendimiento anual (en %): ").strip())
        except ValueError:
            print("⚠️ Entrada inválida. Monto y tasa deben ser números. Registro cancelado.\n")
            continue

        # Guardamos la inversión como diccionario dentro de la lista portafolio
        inversion = {
            "activo": activo_seleccionado,
            "monto_cop": monto_cop,
            "tasa_anual": tasa_anual
        }
        portafolio.append(inversion)
        print(f"✅ Inversión registrada: {activo_seleccionado} - COP {monto_cop:,.2f} - {tasa_anual:.2f}%\n")

    elif opcion == "2":
        # Ver estado del portafolio y comparar con la meta de libertad financiera
        if not portafolio:
            print("\n⚠️ Su portafolio está vacío. Registre al menos una inversión.\n")
            continue

        ingreso_pasivo_mensual_total = 0.0
        print("\nDetalle del portafolio:")
        for inv in portafolio:
            monto = inv["monto_cop"]            # Monto invertido (float)
            tasa = inv["tasa_anual"]           # Tasa anual en porcentaje (float)
            # Fórmula para ingreso pasivo mensual: (Monto * (Tasa / 100)) / 12
            ingreso_mensual = (monto * (tasa / 100.0)) / 12.0
            ingreso_pasivo_mensual_total += ingreso_mensual
            print(f"- {inv['activo']}: COP {monto:,.2f} | Tasa anual {tasa:.2f}% → Ingreso mensual estimado: COP {ingreso_mensual:,.2f}")

        # Resumen y comparación con la meta
        print("\n=================== Resumen ===================")
        print(f"🔸 Ingreso Pasivo Mensual Total estimado: COP {ingreso_pasivo_mensual_total:,.2f}")
        print(f"🎯 Meta de Libertad Financiera (mensual): COP {perfil_usuario['meta_mensual']:,.2f}")

        if ingreso_pasivo_mensual_total >= perfil_usuario["meta_mensual"]:
            print(f"\n🎉 ¡Felicidades {perfil_usuario['nombre']}! Has alcanzado la Libertad Financiera.\n")
        else:
            faltante = perfil_usuario["meta_mensual"] - ingreso_pasivo_mensual_total
            print(f"\n🔻 Aún necesitas: COP {faltante:,.2f} para alcanzar tu meta mensual.\n")

    elif opcion == "3":
        # Salir del programa
        print(f"\nGracias por usar FinVes, {perfil_usuario['nombre']}. ¡Hasta pronto!")
        break

    else:
        # Opción inválida
        print("⚠️ Opción no válida. Seleccione 1, 2 o 3.\n")
```
